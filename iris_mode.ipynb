```python
#import libraries
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
iris_data=load_iris()
#returns a dictionary containing data and some other properties of data
iris_data
```

```python
iris_data['target_names']
```

```python
#look at what data matrix is representing
print(iris_data['feature_names'])
iris_data['data'][:5]
```

```python
#look at number of data points
print(iris_data['data'].shape)
print(iris_data['target'].shape)
```

```python
iris_data['target']
#0 1 2 representing 3 kinds of flowers
```

```python
#split dataset in train and test set (shuffled, stratified, see in documentatio)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris_data['data'],iris_data['target'],random_state=0)
```

```python
# look at train and test set size
print(x_train.shape)
print(x_test.shape)
```

```python
y_train
#randomly shiffled training set
```

```python
#visualizing data to gather more information
#we are using pandas function to create pair plots
iris_dataframe=pd.DataFrame(x_train,columns=iris_data['feature_names'])
iris_dataframe.head()
```

# Plotting pair plots for more info

```python
#draw pair plots
#in pandas scatter_matrix, digonals are histograms
pair_plots=pd.plotting.scatter_matrix(iris_dataframe,alpha=0.8,figsize=(15,15),c=y_train,s=80)
#alpha- transperancy applied
#c- classes 
#s- size of points
```

# Train model

```python
#create a knn classifier object
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
#look at the different parameters which are used in this object
#You don’t have to worry about the other parameters shown in this representation
knn
```

```python
knn.fit(x_train,y_train)
```

# Predict

```python
#lest do prediction on our new instance
x_new=np.array([[5,2.4,1.3,0.3]])
```

```python
prediction=knn.predict(x_new)
print(iris_data['target_names'][prediction])
#our x_new belongs to setosa iris
```

# check Accuracy of model

```python
#use x_test and y_test to compare results
y_pred=knn.predict(x_test)
y_pred
# y_pred giving outputs fror x_test instances
```

```python
test_score=knn.score(x_test,y_test)
print(test_score)
# our model showed 97% accuracy
```
